### Cleaning the data, removing columns which are not needed and removing the duplicates

```python
df.columns
```

```python
df.drop(columns = ['Time', 'Flight no.', 'Registration', 'Schedule', 'MSN', 'YOM', 'Circumstances'], inplace = True)
```

```python
df['Aircraft'] = df['Aircraft'].fillna('unknown')
df['Flight phase'] = df['Flight phase'].fillna('unknown')
df['Flight type'] = df['Flight type'].fillna('unknown')
df['Survivors'] = df['Survivors'].fillna(df['Survivors']).mode()[0]
df['Crash site'] = df['Crash site'].fillna('unknown')
df['Crash location'] = df['Crash location'].fillna('unknown')
df['Country'] = df['Country'].fillna('unknown')
df['Region'] = df['Region'].fillna('unknown')
df['Crew on board'] = df['Crew on board'].fillna(df['Crew on board']).mean()
df['Pax on board'] = df['Pax on board'].fillna(df['Pax on board']).mean()
df['PAX fatalities'] = df['PAX fatalities'].fillna(df['PAX fatalities']).mean() 
df['Crew fatalities'] = df['Crew fatalities'].fillna(df['Crew fatalities']).mean()
df['Other fatalities'] = df['Other fatalities'].fillna(df['Other fatalities']).mean()
```

```python
df.isnull().any()
```

### Converting the data types from objects to its corresponding data types

```python
df[df.select_dtypes(include = 'object').columns] = df.select_dtypes(include = 'object').astype('category') 
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
```

```python
df['Crew on board'] = pd.to_numeric(df['Crew on board'], errors = 'coerce').astype('int64')
df['Crew fatalities'] = pd.to_numeric(df['Crew fatalities'], errors = 'coerce').astype('int64')
df['Pax on board'] = pd.to_numeric(df['Pax on board'], errors = 'coerce').astype('int64')
df['PAX fatalities'] = pd.to_numeric(df['PAX fatalities'], errors = 'coerce').astype('int64')
df['Other fatalities'] = pd.to_numeric(df['Other fatalities'], errors = 'coerce').astype('int64')
```

### Analyzing based on our requirements

### 1. How has the number of airplane crashes changed over the decades? 

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Decade'] = (df['Date'].dt.year // 10) * 10
DecadeCrash = df.groupby('Decade').size()
print(DecadeCrash)
```

### 2. What is the total number of crashes per year?

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
totalCrashPerYear = df.groupby('Year')['Date'].size().head(10)
data = pd.DataFrame({'Total crashes for the first 10 years': totalCrashPerYear})
data
```

### 3. Top 10 highest crashes of all time 

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
HighestCrashPerYear = df.groupby('Year')['Date'].size().sort_values(ascending = False).head(10)
output = pd.DataFrame({ 'Top 10 highest crashes : ' : HighestCrashPerYear})
output
```

### 4. Which 5 aircraft have had the most airplane crashes?

```python
AircraftsWithMostNumberOfCrashes = df.groupby('Aircraft',observed = True)['Total fatalities'].size().sort_values(ascending = False).head(5)
AircraftsWithMostNumberOfCrashes
```

### 5. Which 3 countries have recorded the most airplane crashes?


```python
MostAffectedCountry = df.groupby('Country', observed = True)['Country'].size().sort_values(ascending = False).head(3)
MostAffectedCountry
```

### 6. Most Dangerous Aircraft Model

```python
MostDangerousAircraftModel = df.groupby('Aircraft', observed = True)['Aircraft'].size().sort_values(ascending = False).head(1)
MostDangerousAircraftModel
```

### 7. Which month has had the most number of crashes 

```python
df['Month'] = pd.to_datetime(df['Date']).dt.month
MonthWithHighestCrashes = df.groupby('Month', observed = True)['Date'].count()
print('Total Crashes in each month : ',MonthWithHighestCrashes)
```

```python
df['Month'] = pd.to_datetime(df['Date']).dt.month
MonthWithHighestCrashes = df.groupby('Month', observed=True)['Date'].size().sort_values(ascending=False)

highest_month = MonthWithHighestCrashes.idxmax()  
highest_count = MonthWithHighestCrashes.max()  

print(f"The month with the most number of crashes is Month {highest_month} with {highest_count} crashes.")
```

### 8. Crew on board vs Crew fatalities

```python
# Crew on board vs Crew fatalities
CrewOnBoard = df['Crew on board'].sum()
CrewFatalities = df['Crew fatalities'].sum()
TotalSurvivors = CrewOnBoard - CrewFatalities
FatalityRate = (CrewFatalities / CrewOnBoard) * 100
FatalityRate = FatalityRate.round(2)  
data = pd.Series({
    'Total number of Crew on board': CrewOnBoard,
    'Total number of Crew fatalities': CrewFatalities,
    'Total survivors': TotalSurvivors,
    'Fatality Rate (%)': FatalityRate 
})
data
```

```python
# Top 10 aircraft crashes Crew on board vs Crew fatalities
CrewOnBoard = df.groupby('Aircraft', observed = False)['Crew on board'].sum().sort_values(ascending = False).head(10)
CrewFatalities = df.groupby('Aircraft', observed = False)['Crew fatalities'].sum().sort_values(ascending = False).head(10)
data = pd.DataFrame({'Crew On Board ' : CrewOnBoard, 'Crew fatalities' : CrewFatalities})
data
```

### 9. PAX on board vs PAX fatalities

```python
# Passengers on board vs Passengers fatalities
PaxOnBoard = df['Pax on board'].sum()
PaxFatalities = df['PAX fatalities'].sum()
TotalSurvivors = PaxOnBoard - PaxFatalities
FatalityRate = (PaxFatalities / PaxOnBoard) * 100
FatalityRate = FatalityRate.round()

data = pd.Series({
    'Total number of Passengers on board': PaxOnBoard,
    'Total number of Passenger fatalities': PaxFatalities,
    'Total survivors': TotalSurvivors,
    'Fatality Rate (%)': FatalityRate
})
data
```

```python
# Top 10 aircraft crashes Crew on board vs Crew fatalities
PAXOnBoard = df.groupby('Aircraft', observed = False)['Pax on board'].sum().sort_values(ascending = False).head(10)
PAXFatalities = df.groupby('Aircraft', observed = False)['PAX fatalities'].sum().sort_values(ascending = False).head(10)
data = pd.DataFrame({'Pax on board ' : PAXOnBoard, 'PAX fatalities' : PAXFatalities})
data
```

### 10. Most Crash Cause

```python
crashCause = df.groupby('Crash cause', observed = False)['Crash cause'].count()
totalCrash = df['Crash cause'].count()
print('Total crashes occured so far :',totalCrash)
print('The various causes of the crash occured are : \n',crashCause)
```

### 11. Top 10 crash location 

```python
CrashLocation = df.groupby('Crash location', observed = False)['Total fatalities'].sum().sort_values(ascending = False).head(10)
print('The various crash locations are : \n',CrashLocation)
```

### 12. Most common Crash sites

```python
CrashSite = df.groupby('Crash site', observed = False )['Total fatalities'].sum()
print('The most common crash sites are : \n',CrashSite)
```

### 13. What is the survival rate for passengers and crew across different crash causes?

```python
Survival = df.groupby('Crash site', observed = False ).agg(
    { 'PAX fatalities': 'sum', 'Crew fatalities': 'sum', 'Other fatalities': 'sum', 'Total fatalities': 'sum'}
)
Survival
```

### 14. Which flight phase has had the most number of crashes

```python
TotalFatalitiesDuringPhase = df.groupby('Flight phase', observed = False).agg(
    { 'PAX fatalities': 'sum', 'Crew fatalities': 'sum', 'Total fatalities': 'sum'}
)
TotalFatalitiesDuringPhase
```

### 15. 10 highest fatalities of each decade 

```python
country_decade_stats = df.groupby(['Country', 'Decade'], observed = False).agg(
    Total_Crashes=('Aircraft', 'size'),
    Total_Fatalities=('Total fatalities', 'sum'))
country_decade_stats_sorted = country_decade_stats.reset_index().sort_values(by='Total_Crashes', ascending=False).head(10)
country_decade_stats_sorted
```
